## Feature engineering

### Subtask:
Create relevant liquidity-related features such as moving averages, volatility measures, liquidity ratios, and potentially sentiment indicators.


```python
df_crypto_cleaned['MA_14'] = df_crypto_cleaned['close'].rolling(window=14).mean()
df_crypto_cleaned['MA_50'] = df_crypto_cleaned['close'].rolling(window=50).mean()
df_crypto_cleaned['volatility'] = df_crypto_cleaned['close'].rolling(window=14).std()
df_crypto_cleaned['liquidity_ratio'] = (df_crypto_cleaned['close'] - df_crypto_cleaned['open']) / df_crypto_cleaned['volume']

display(df_crypto_cleaned.head())
display(df_crypto_cleaned.isnull().sum())
```

## Model selection and training

### Subtask:
Choose appropriate machine learning models for time-series forecasting or regression (e.g., ARIMA, LSTM, Prophet, Linear Regression, Random Forest). Split data into training and testing sets. Train the selected model(s) on the training data.


```python
X = df_crypto_cleaned.drop(['timestamp', 'traded_value', 'MA_14', 'MA_50', 'volatility', 'liquidity_ratio'], axis=1)
y = df_crypto_cleaned['traded_value']

split_ratio = 0.8
split_index = int(len(df_crypto_cleaned) * split_ratio)

X_train = X[:split_index]
y_train = y[:split_index]
X_test = X[split_index:]
y_test = y[split_index:]

from sklearn.ensemble import RandomForestRegressor

model = RandomForestRegressor(n_estimators=100, random_state=42)

model.fit(X_train, y_train)
```

```python
from sklearn.metrics import mean_absolute_error, r2_score

y_pred = model.predict(X_test)

mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Absolute Error: {mae}")
print(f"R-squared: {r2}")
```
